# Data cleaning 
* **Here we are removing all uneccessary columns.**


```python
movies = movies.drop(['id','homepage','keywords','original_language','overview','production_companies',
'spoken_languages','original_title','vote_count','tagline'], axis = 1) 
```

```python
#displaying available left
movies.shape
```

#### CHECKING DATA TYPE  AND NULL COUNT

```python
movies.info()
```

#### CHECKING FOR NULL VALUES

```python
movies.isnull().sum()
```

#### Dropping null field

###### From the display above we need to drop the missing values for release date and runtime to help run a smooth analysis.

```python

movies = movies.dropna(subset=['runtime','release_date'])
movies.shape
```

```python
#Checking all movie information after dropping null values

movies.isnull().sum()
movies.info()
```

```python
#### Changing format of budget and revenue column from float to int type

change_type=['budget', 'revenue']
movies[change_type]=movies[change_type].applymap(np.int64)
movies.info()
```

#### Removing all rows with 0 in both budget and revenue columns respectively

```python
temp_list=['budget', 'revenue']

#this replaces all the value from '0' to NAN in the list
movies[temp_list] = movies[temp_list].replace(0, np.NAN)

#Removing all the rows with NaN value in temp_list 
movies.dropna(subset = temp_list, inplace = True)
```

# Creating a new column for profit

###### Since we will be finding movies with higher profit, we need to insert a new columns called profit_earned, mathematically we can find profit_earned by illustrating that ; profit_earned = revenue - budget since we have these parameters in our datasets.

```python
budget = movies['budget']
revenue = movies['revenue']

movies.insert(2,'profit_earned',revenue-budget)
```

```python
movies.head()
```

###### From the above table dispaly, we can see that profit_earned column has been added  to table. For easy way to find highly grossed movies in our analysis.

#### Function for line plotter

```python
def line_plotter(x_axis,y_axis):
    plotter_var = movies.groupby(x_axis)[y_axis].sum()
    #figure size(width, height)
    plt.figure(figsize=(10,6))

    #x-axis
    plt.xlabel(x_axis, fontsize = 12)
    #y-axis
    plt.ylabel(y_axis, fontsize = 12)
    #title
    plt.title(f'Total {y_axis} by all movies vs {x_axis} of the movies.')

    #plotting the graph
    plt.plot(plotter_var )

    #displaying the line plot
    plt.show()
```

#### How long were the highest grossing movies?

To find out how long were the highest grossing movies,we have can plot a profit_earned against runtime of all the movies.
