# EXPLORING FILMS OVER SOME YEARS

# In this project, we will analyze a dataset on movies collection over some past years.


```python
#import modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read movies 
movies = pd.read_csv("/home/nitro/Documents/archive/tmdb_5000_movies.csv")

#Print the first five rows
movies.head()
```

# DATA CLEANING (Getting rid of all unnecessary data )

```python
#dropping homepage since it is not is neccessary in my analysis.
movies = movies.drop("homepage", axis = 1)
```

```python
movies.head()
```

```python
# dropping production countries column
movies = movies.drop("production_countries",axis = 1)
```

```python
movies.head()
```

```python
#dropping tagline
movies= movies.drop("tagline", axis =1)
```

```python
movies.head()
```

```python
#dropping original language
movies = movies.drop("original_language", axis =1)
```

```python
movies.head()
```

```python
#dropping Status
movies = movies.drop('status',axis =1)
```
